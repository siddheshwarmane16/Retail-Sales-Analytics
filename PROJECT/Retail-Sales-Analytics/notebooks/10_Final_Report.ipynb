```python
# ==========================================================
# Cell 1 : Import Libraries
# ==========================================================

import os
import pandas as pd

from reportlab.lib import colors
from reportlab.lib.units import cm
from reportlab.lib.pagesizes import A4

from reportlab.platypus import (
    SimpleDocTemplate,
    Paragraph,
    Spacer,
    Image,
    Table,
    TableStyle
)

from reportlab.lib.styles import getSampleStyleSheet

# Create reports folder
os.makedirs("../reports", exist_ok=True)

# Load dataset
df = pd.read_csv(
    "../data/processed/retail_clean.csv",
    parse_dates=["Order_Date"]
)

print("Dataset Loaded Successfully")
print(df.shape)
```

```python
# ==========================================================
# Cell 2 : Business KPIs
# ==========================================================

total_sales = df["Sales"].sum()

total_profit = df["Profit"].sum()

total_orders = df["Order_ID"].nunique()

avg_order_value = total_sales / total_orders

top_category = (
    df.groupby("Category")["Sales"]
      .sum()
      .idxmax()
)

top_region = (
    df.groupby("Region")["Profit"]
      .sum()
      .idxmax()
)

kpi_data = [
    ["Metric", "Value"],
    ["Total Sales", f"₹ {total_sales:,.2f}"],
    ["Total Profit", f"₹ {total_profit:,.2f}"],
    ["Total Orders", f"{total_orders:,}"],
    ["Average Order Value", f"₹ {avg_order_value:,.2f}"],
    ["Top Category", top_category],
    ["Top Region", top_region]
]

display(pd.DataFrame(kpi_data[1:], columns=kpi_data[0]))
```

```python
# ==========================================================
# Cell 3 : Create PDF
# ==========================================================

styles = getSampleStyleSheet()

doc = SimpleDocTemplate(
    "../reports/Final_Business_Report.pdf",
    pagesize=A4
)

story = []

story.append(
    Paragraph(
        "Retail Sales Analytics Report",
        styles["Title"]
    )
)

story.append(Spacer(1,15))

story.append(
    Paragraph(
        "Executive Summary",
        styles["Heading2"]
    )
)

story.append(
    Paragraph(
        """
        This report summarizes sales performance,
        customer behaviour,
        forecasting,
        recommendation engine,
        and machine learning model results.
        """,
        styles["BodyText"]
    )
)

story.append(Spacer(1,15))
```

```python
# ==========================================================
# Cell 4 : KPI Table
# ==========================================================

table = Table(
    kpi_data,
    colWidths=[8*cm,7*cm]
)

table.setStyle(
    TableStyle([

        ("BACKGROUND",(0,0),(-1,0),colors.HexColor("#1565C0")),

        ("TEXTCOLOR",(0,0),(-1,0),colors.white),

        ("GRID",(0,0),(-1,-1),0.5,colors.grey),

        ("BACKGROUND",(0,1),(-1,-1),colors.whitesmoke),

        ("FONTNAME",(0,0),(-1,0),"Helvetica-Bold"),

        ("BOTTOMPADDING",(0,0),(-1,0),10),

        ("ALIGN",(0,0),(-1,-1),"CENTER")

    ])
)

story.append(table)

story.append(Spacer(1,20))
```

```python
# ==========================================================
# Cell 5 : Add Charts
# ==========================================================

charts = [

("../images/sales_by_category.png",
"Sales by Category"),

("../images/regional_performance.png",
"Regional Performance"),

("../images/model_comparison.png",
"Machine Learning Models"),

("../images/customer_segments.png",
"Customer Segmentation"),

("../images/sales_forecast_prophet.png",
"Sales Forecast")

]

for image_path, title in charts:

    if os.path.exists(image_path):

        story.append(
            Paragraph(
                title,
                styles["Heading2"]
            )
        )

        story.append(
            Image(
                image_path,
                width=16*cm,
                height=9*cm
            )
        )

        story.append(
            Spacer(1,15)
        )

    else:

        print(f"Image Missing : {image_path}")
```

```python
# ==========================================================
# Cell 6 : Business Insights
# ==========================================================

story.append(
    Paragraph(
        "Business Insights",
        styles["Heading2"]
    )
)

story.append(
    Paragraph(
        f"""

• Total Revenue Generated : ₹ {total_sales:,.2f}<br/>

• Total Profit : ₹ {total_profit:,.2f}<br/>

• Best Performing Category : {top_category}<br/>

• Highest Profit Region : {top_region}<br/>

• Recommendation Engine Built using Cosine Similarity.<br/>

• Customer Segmentation performed using K-Means Clustering.<br/>

• Sales Forecasting performed using Facebook Prophet.<br/>

• Machine Learning models evaluated using R², RMSE and MAE.

        """,

        styles["BodyText"]

    )
)
```

```python
# ==========================================================
# Cell 7 : Generate Report
# ==========================================================

doc.build(story)

print("="*60)
print("FINAL BUSINESS REPORT GENERATED SUCCESSFULLY")
print("="*60)

print("Location")

print("../reports/Final_Business_Report.pdf")
```
