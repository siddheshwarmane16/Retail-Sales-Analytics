```python
# ============================================
# Cell 1: Import Required Libraries
# ============================================

import os
from datetime import datetime, timedelta

import numpy as np
import pandas as pd

# Set random seed for reproducibility
np.random.seed(42)
```

```python
# ============================================
# Cell 2: Generate Synthetic Retail Dataset
# ============================================

# Dataset Parameters
n_rows = 20000
n_customers = 1500
n_products = 120

regions = ["North", "South", "East", "West", "Central"]

categories = [
    "Electronics",
    "Clothing",
    "Grocery",
    "Furniture",
    "Toys",
    "Sports",
    "Beauty"
]

sub_categories = {
    "Electronics": ["Mobiles", "Laptops", "Accessories"],
    "Clothing": ["Men", "Women", "Kids"],
    "Grocery": ["Beverages", "Snacks", "Dairy"],
    "Furniture": ["Chairs", "Tables", "Sofas"],
    "Toys": ["Educational", "Outdoor", "Puzzles"],
    "Sports": ["Fitness", "Outdoor Games", "Apparel"],
    "Beauty": ["Skincare", "Haircare", "Makeup"]
}

# Generate Random Dates
start_date = datetime(2022, 1, 1)

date_range = [
    start_date + timedelta(days=int(x))
    for x in np.random.randint(0, 1095, n_rows)
]

# Create DataFrame
df = pd.DataFrame({
    "Order_ID": [f"ORD{100000+i}" for i in range(n_rows)],
    "Order_Date": date_range,
    "Customer_ID": [
        f"CUST{np.random.randint(1, n_customers):05d}"
        for _ in range(n_rows)
    ],
    "Product_ID": [
        f"PROD{np.random.randint(1, n_products):04d}"
        for _ in range(n_rows)
    ],
    "Category": np.random.choice(categories, n_rows),
    "Region": np.random.choice(regions, n_rows),
    "Quantity": np.random.randint(1, 10, n_rows),
    "Unit_Price": np.round(np.random.uniform(5, 500, n_rows), 2),
    "Discount": np.round(np.random.uniform(0, 0.30, n_rows), 2)
})

# Assign Sub-Category
df["Sub_Category"] = df["Category"].apply(
    lambda x: np.random.choice(sub_categories[x])
)

# Calculate Sales
df["Sales"] = np.round(
    df["Quantity"] * df["Unit_Price"] * (1 - df["Discount"]),
    2
)

# Calculate Cost
df["Cost"] = np.round(
    df["Sales"] * np.random.uniform(0.55, 0.80, n_rows),
    2
)

# Calculate Profit
df["Profit"] = np.round(
    df["Sales"] - df["Cost"],
    2
)

# Introduce Missing Values
missing_index = np.random.choice(df.index, 300, replace=False)
df.loc[missing_index, "Unit_Price"] = np.nan

# Introduce Duplicate Records
duplicates = df.sample(150, random_state=42)
df = pd.concat([df, duplicates], ignore_index=True)

print("Dataset Generated Successfully!")
print("Shape:", df.shape)
```

```python
# ============================================
# Cell 3: Save Dataset
# ============================================

# Create folder if it doesn't exist
os.makedirs("../data/raw", exist_ok=True)

# Save dataset
file_path = "../data/raw/retail_transactions.csv"

df.to_csv(file_path, index=False)

print(f"Dataset saved successfully at:\n{file_path}")
```
